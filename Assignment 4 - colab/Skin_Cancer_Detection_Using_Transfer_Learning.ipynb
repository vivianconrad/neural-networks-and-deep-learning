```python
data = pd.read_csv('HAM10000_metadata.csv')
# Set y as the labels
y = data['dx']
```

```python
#split data
from sklearn.model_selection import train_test_split
data_train, data_value = train_test_split(data, test_size=0.1, random_state=101, stratify=y)
```

```python
# Transfer the images into folders, Set the image id as the index
image_index = data.set_index('image_id', inplace=True)
```

```python
# Get a list of images in each of the two folders
folder_1 = os.listdir('HAM10000_images_part_1')
folder_2 = os.listdir('HAM10000_images_part_2')
```

```python
# Get a list of train and val images
train_list = list(data_train['image_id'])
value_list = list(data_value['image_id'])
```

```python
for img in train_list:
  file_name = img + '.jpg'
  if file_name in folder_1:
          #the source path
          source = os.path.join('HAM10000_images_part_1', file_name)
          
          #the destination path
          destination = os.path.join(train_dir+'/'+data['dx'][img], file_name)
          print(destination)
          
          shutil.copyfile(source, destination)
  if file_name in folder_2:
          #the source path
          source = os.path.join('HAM10000_images_part_2', file_name)
          #the destination path
          destination = os.path.join(train_dir, file_name)
          
          shutil.copyfile(source, destination)
```

```python
from keras.preprocessing.image import ImageDataGenerator
import keras 
```

```python
print(data.head())
img_class = ['nv','mel','bkl','bcc','akiec','vasc','df']
train_path = 'base_dir/train_dir/'
valid_path = 'base_dir/val_dir/'
```

```python
print(os.listdir('base_dir/train_dir'))
print(len(os.listdir('base_dir/val_dir')))
img_shape = 224
train_data_gen  = ImageDataGenerator(rescale=1./255)
value_data_gen  = ImageDataGenerator(rescale=1./255)
```

```python
#declares data generator for train and val batches
batches = train_data_gen.flow_from_directory(train_path, 
                                                        target_size = (img_shape,img_shape),
                                                        classes = img_class,
                                                        batch_size = 64
                                                        )
valid_batches = value_data_gen.flow_from_directory(valid_path, 
                                                        target_size = (img_shape,img_shape),
                                                        classes = img_class,
                                                        batch_size = 64                                                      )
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten 
from keras.callbacks import ReduceLROnPlateau
from keras.models import Model
```

```python
mobiles = keras.applications.mobilenet.MobileNet()
a = mobiles.layers[-6].output
```

```python
# Add a dropout and dense layer for predictions
a = Dropout(0.25)(a)
predictions = Dense(7, activation='softmax')(a)
print(mobiles.input)
net = Model(inputs=mobiles.input, outputs=predictions)
mobiles.summary()
```

```python
for layer in net.layers[:-23]:
  layer.trainable = False
net.compile(optimizer='adam',
  loss='categorical_crossentropy',
  metrics=['accuracy'])
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', patience=3, verbose=1, factor=0.5, min_lr=0.00001)
history = net.fit_generator(batches, epochs=10)
```
