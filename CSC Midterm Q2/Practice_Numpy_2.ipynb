## Create a boolean array

```
import numpy as np
a1 = np.array([1, 0, 1], dtype='bool')
print(a1)
```

## Create an object array to hold numbers as well as strings

```
a2 = np.array([1, 'a'], dtype='object')
print(a2)
print(type(a2))
```

## Convert an array back to a list

```
a2 = np.array([1, 'a'], dtype='object')
a3=a2.tolist()
print(a3)
print(type(a3))
```

## Represent missing values

```
a4 = np.array([[1, 'sas', 3],
               [4, 5, 6]])
print(a4)
print('\n')

a4[0,1] = np.nan  # not a number
print(a4)
```

## Compute mean, min, max on the ndarray

```
a5 = np.random.randint(10, size=(4, 5))
print(a5)
print('\n')
print("Mean value is: ", a5.mean())
print("Max value is: ", a5.max())
print("Min value is: ", a5.min())
```

## Compute the minimum values row wise or column wise

```
a6 = np.random.randint(10, size=(4, 5))
print(a6)
print('\n')
print("Column wise minimum: ", np.amin(a6, axis=0))
print("Row wise minimum: ", np.amin(a6, axis=1))
```

## Compute cumulative sum

```
a7 = np.random.randint(10, size=(4, 5))
print(a7)
print('\n')
print(np.cumsum(a7))
```

## Obtain the common items between two numpy arrays

```
a8 = np.array([1,2,3,2,3,4,3,4,5,6])
a9 = np.array([7,2,10,2,7,4,9,4,9,8])

a10=np.intersect1d(a8,a9)
print(a10)
```

## Remove items from one numpy array that exist in another


```
a11 = np.array([1,2,3,4,5])
a12 = np.array([5,6,7,8,9])

a13 = np.setdiff1d(a11,a12)
print(a13)
```

## Obtain the positions where elements of two numpy arrays match

```
a14 = np.array([1,2,3,2,3,4,3,4,5,6])
a15 = np.array([7,2,10,2,7,4,9,4,9,8])

a16=np.where(a14 == a15)
print(a16)
```

## Q1: Extract all numbers between between 5 and 15 from x

- **Input** x = np.array([2, 6, 1, 9, 10, 3, 27])
- **Expected Output** [ 6  9 10]


```

x = np.array([2, 6, 1, 9, 10, 3, 27])
y = np.where((x >= 5) & (x <= 15))
z=x[y]
print(z)
```
