```python
# MLP for the I MDB problem
# importing required libraries
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
```

```python
# loading the only top (n) words from the dataset 
top_words = 5000
(train_data_x, train_data_y), (test_x, test_y) = imdb.load_data(num_words=top_words)
# zero-padding and truncating the dataset to a maximum review length in words
max_words = 500
train_data_x = sequence.pad_sequences(train_data_x, maxlen=max_words)
test_x = sequence.pad_sequences(test_x, maxlen=max_words)
```

```python
# creating the model
model = Sequential()
model.add(Embedding(top_words, 32, input_length=max_words))
model.add(Flatten())
model.add(Dense(250, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Fitting the model over 5 epochs
model.fit(train_data_x, train_data_y, validation_data=(test_x, test_y), epochs=5, batch_size=128, verbose=2)
```

```python
# Final evaluation of the model
scores = model.evaluate(test_x, test_y, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
